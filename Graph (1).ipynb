```python
import networkx as nx#used to be able to create graph


G=nx.Graph()#we create our empty graph
G.add_nodes_from(range(1,11))#give it a node range, creating 10 nodes
G.add_edges_from([(1,2),(2,3),(3,4),(3,5),(4,6),(5,7),(6,9),(7,10),(4,8),(9,8),(10,8),(5,8)])#we connect nodes in desired order
G.add_weighted_edges_from([(4,5,.5),(9,10,.5),(4,9,.25),(5,10,.25),(6,8,.35),(7,8,.35)]) #some nodes are further away plus helps create the shape


nx.draw(G, with_labels = True)#draw graph G with labes
```
