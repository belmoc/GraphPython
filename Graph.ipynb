```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()
G.add_nodes_from(range(1,11))
G.add_edges_from([(1,2),(2,3),(3,4),(3,5),(4,6),(5,7),(6,9),(7,10),(4,8),(9,8),(10,8),(5,8)])
G.add_weighted_edges_from([(4,5,.5),(9,10,.5),(4,9,.25),(5,10,.25),(6,8,.35),(7,8,.35)])


nx.draw(G, with_labels = True)
plt.draw()
plt.show()
```
